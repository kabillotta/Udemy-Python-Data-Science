```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1.,1.])
axes2 = fig.add_axes([0.2, 0.7, .3,.3])
axes.plot(range(0,100), np.random.rand(100), 'b')
axes2.plot(range(0,100), np.random.rand(100), 'r')
axes.plot(range(0,100), np.random.randn(100), 'g')
axes.set_xlabel('Define o Label X') 
axes.set_ylabel('Define o Label Y')
axes.set_title('Define o TÃ­tulo')
```
